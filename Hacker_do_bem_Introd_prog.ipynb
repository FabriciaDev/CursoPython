# 4. Estrutura de Repetição

Para que um conjunto de determinadas instruções seja executado repetidamente em Python, são empregadas as estruturas de repetição. Também conhecidas como loops, são componentes essenciais em programação quando uma tarefa precisa ser executada repetidas vezes

## 4.1 While

While é um loop em Python que realiza o controle de fluxo que permite a execução repetida de um bloco de código enquanto uma condição específica permanece verdadeira.

Vamos criar um programa que pergunta ao usuário um número inteiro qualquer  e o programa calcula automaticamente a tabuada de 1 até 10 desse número
* Sem o uso do comando **while**

```python
# Recebendo o número do usuário
print('='*43)
numero = input("digite um númro para o cálculo da sua tabuada \n")
print('='*43)
# Coonvertendo o número inteiro
numero = int(numero)
# Exibindo a tabuada do número
print(f"{numero} x 1 = {numero*1}")
print(f"{numero} x 2 = {numero*2}")
print(f"{numero} x 3 = {numero*3}")
print(f"{numero} x 4 = {numero*4}")
print(f"{numero} x 5 = {numero*5}")
print(f"{numero} x 6 = {numero*6}")
print(f"{numero} x 7 = {numero*7}")
print(f"{numero} x 8 = {numero*8}")
print(f"{numero} x 9 = {numero*9}")
print(f"{numero} x 10 = {numero*10}")
```

* Com o uso do comando **while**

```python
# Recebendo o número do usuário
print('='*45)
numero = input("digite um númro para o cálculo da sua tabuada \n")
print('='*45)
# Coonvertendo o número inteiro
numero = int(numero)
# Criando a variável de controle de floxo
contador = 1
while(contador<=10):
  #Exibindo o valor da tabuada
  print(f"{numero} X {contador} = {numero*contador}")
  contador = contador + 1
print('='*45)
print("FIM DE PROGRAMA")
print('='*45)
```

## 4.2 Acumuladores

Para esse exemplo vamos criar um programa que calcula a média das idades de 5 pessoas

Podemos, nos laços de repetição, ter a necessidade de somar valores a cada repetição dos mesmos. Para isso podemos usar uma variável acumulador que irá acumulando esses valores
* Sem o uso de acumuladores:

```python
# REcebendo as idades dos usuários
print('='*43)
id1 = input("digite a primeira idade:\n")
print('='*43)
id2 = input("digite a segunda idade:\n")
print('='*43)
id3 = input("digite a terceira idade:\n")
print('='*43)
id4 = input("digite a quarta idade:\n")
print('='*43)
id5 = input("digite a quinta idade:\n")
print('='*43)
# Convertendo para números inteiros
id1 = int(id1)
id2 = int(id2)
id3 = int(id3)
id4 = int(id4)
id5 = int(id5)
# Calculando a média
media = int((id1 + id2 + id3 + id4 + id5)/5)
# Exibindo a média das idades
print('='*43)
print(f"A média das idades é igual a {media}")
print('='*43)
print('='*43)
```

* Com o uso de **while**

```python
# Criando a variável de controle de laço, contando até 5
print('='*43)
print('='*43)
contador = 1
# Criando a variável para acumular os valores das idades
soma = 0
# Criando o laço de repetiçao para receber
while(contador<=5):
  # Recebendo a idade do usuário
  idade = int(input(f"Digite a idade do usuário número {contador}: "))
  print('='*43)
  # Somando a idade do usuário ao valor
  soma = soma + idade
  # Incrementando variável contador para controle de loop
  contador = contador + 1
# Calculando a média das idades
media = soma/(contador-1)
print(f"A média das idades digitadas é igual: {media}!!!")
print('='*43)
print("FIM")
print('='*43)
print('='*43)32
```

## 4.3 Comando For

O comando For é uma estrutura de controle de fluxo que facilita a iteração sobre uma sequência de elementos e é utilizado quando o número de iterações é conhecido antecipadamente.

image.png

```python
# range (inicio, final,incremento)
for cont in range(0,10,2):
  print(cont)
```

```python
for i in range(10,-3,-1):
  print(i)
```

```python
# Entrada de dados:
num = int(input("Digite um número para exibir a tabuada:\n"))
# Laço de repetição
for i in range(1,11):
  print(f"{num} X {i} = {num*i}")
```
